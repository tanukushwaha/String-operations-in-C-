```
from matplotlib import pyplot as plt
x=[5,8,1]
y=[12,13,14]

plt.plot(x,y)
plt.title("Epic Info")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()
```

```
from matplotlib import pyplot as plt
from matplotlib import style
style.use("ggplot")
x1=[5,8,10]
y1=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x1,y1,label="line one",linewidth=5,color="green")
plt.plot(x2,y2,linewidth=5)
plt.title("Epic Info")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.legend()
plt.grid(True,color="k")
plt.show()
```

```
from matplotlib import pyplot as plt
from matplotlib import style
style.use("ggplot")
x1=[5,8,10]
y1=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x1,y1,label="line one",linewidth=5,color="green")
plt.plot(x2,y2,linewidth=5)
plt.title("Epic Info")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.legend()
plt.grid(True,color="k")
plt.bar(x1,y1,align="center")
plt.bar(x2,y2,align="center")
plt.scatter(x1,y1)
plt.show()
```
